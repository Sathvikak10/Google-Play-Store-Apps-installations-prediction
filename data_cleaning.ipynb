```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_csv('Google-Playstore.csv')
```

```python
print(dataset.describe())
```

```python
dataset.dropna(inplace=True)
```

```python
print(dataset.count())
```

```python
print(dataset.isna().sum())
```

```python
print(dataset.dtypes)
```

```python
dataset.Rating = dataset.Rating.astype('float')
```

```python
print(dataset.Size.head())
```

```python
dataset.Size = dataset.Size.str.replace(',', '')
dataset.Size = dataset.Size.apply(lambda x: float(x[:-1]) * 0.001 if 'k' in x else x)

dataset.Size = dataset.Size.str.replace('M', '')
dataset.Size = dataset.Size.str.replace('G', '')
dataset.Size = dataset.Size.replace('Varies with device', np.NaN)
dataset.Size = dataset.Size.astype('float')
```

```python
print(dataset.groupby('Category')['Size'].mean())
```

```python
dataset = dataset.drop(['Last Updated', 'App Id', 'Minimum Installs', 
                       'Maximum Installs', 'Free', 'Currency', 'Minimum Android', 'Developer Id', 
                       'Developer Website', 'Developer Email', 'Released', 'Privacy Policy',
                       'Ad Supported', 'In App Purchases', 'Editors Choice', 'Scraped Time'], axis=1)
```

```python
travel_index = dataset[dataset['Category'] == 'TRAVEL'].index.values
print(travel_index)
dataset.loc[travel_index, 'Category'] = 'TRAVEL_AND_LOCAL'
```

```python
print(dataset[dataset['Size'].isna()].index.values)
print(dataset.Size)
print(dataset.Size.isna().sum())
dataset['Size'] = dataset.groupby('Category')['Size'].transform(lambda x: x.fillna(x.mean()))
```

```python
print(dataset.Installs.value_counts())
```

```python
dataset.loc[dataset['Installs'].isin(['0+', '1+', '5+', '10+', '50+']), 'Installs'] = '0 - 100'
dataset.loc[dataset['Installs'] == '100+', 'Installs'] = '100 - 500'
dataset.loc[dataset['Installs'] == '500+', 'Installs'] = '500 - 1,000'
dataset.loc[dataset['Installs'] == '1,000+', 'Installs'] = '1,000 - 5,000'
dataset.loc[dataset['Installs'] == '5,000+', 'Installs'] = '5,000 - 10,000'
dataset.loc[dataset['Installs'] == '10,000+', 'Installs'] = '10,000 - 50,000'
dataset.loc[dataset['Installs'] == '50,000+', 'Installs'] = '50,000 - 100,000'
dataset.loc[dataset['Installs'] == '100,000+', 'Installs'] = '100,000 - 500,000'
dataset.loc[dataset['Installs'] == '500,000+', 'Installs'] = '500,000 - 1,000,000'
dataset.loc[dataset['Installs'] == '1,000,000+', 'Installs'] = '1,000,000 - 5,000,000'
dataset.loc[dataset['Installs'] == '5,000,000+', 'Installs'] = '5,000,000 - 10,000,000'
dataset.loc[dataset['Installs'] == '10,000,000+', 'Installs'] = '10,000,000 - 50,000,000'
dataset.loc[dataset['Installs'] == '50,000,000+', 'Installs'] = '50,000,000 - 100,000,000'
dataset.loc[dataset['Installs'] == '100,000,000+', 'Installs'] = '100,000,000 - 500,000,000'
dataset.loc[dataset['Installs'] == '500,000,000+', 'Installs'] = '500,000,000 - 1,000,000,000'
dataset.loc[dataset['Installs'] == '1,000,000,000+', 'Installs'] = '1,000,000,000 - 5,000,000,000'
dataset.loc[dataset['Installs'] == '5,000,000,000+', 'Installs'] = '5,000,000,000+'
```

```python
dataset.Installs = pd.Categorical(dataset.Installs,
                                      ['0 - 100', '100 - 500', '500 - 1,000', '1,000 - 5,000', '5,000 - 10,000',
                                       '10,000 - 50,000', '50,000 - 100,000', '100,000 - 500,000',
                                       '500,000 - 1,000,000', '1,000,000 - 5,000,000', '5,000,000 - 10,000,000',
                                       '10,000,000 - 50,000,000', '50,000,000 - 100,000,000',
                                       '100,000,000 - 500,000,000', '500,000,000 - 1,000,000,000',
                                       '1,000,000,000 - 5,000,000,000', '5,000,000,000+'])
```

```python
print(dataset.Installs.value_counts().sort_index())
```

```python
games = dataset['Category'].str.contains('GAME')
dataset['Game_genre'] = dataset.loc[games, 'Category']
print(dataset['Game_genre'])
dataset['Game_genre'] = dataset['Game_genre'].str.replace('GAME_', '')
```

```python
dataset.loc[games, 'Category'] = 'GAME'
```

```python
print(dataset.Game_genre.count())
print(dataset.Category.value_counts())
```
